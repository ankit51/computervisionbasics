### Import Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

### Read and View Image

```python
image = plt.imread("../input/dog1.jpeg")
plt.imshow(image)
```

### RGB to Grayscale

```python
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

### Manual Thresholding

```python
(T, threshInv) = cv2.threshold(gray_img, 250, 255, cv2.THRESH_BINARY_INV)
masked = cv2.bitwise_and(image, image, mask=threshInv)
plt.imshow(masked)
print("Threshold: ", T)
```

### Triangle Thresholding

```python
(T, threshInv) = cv2.threshold(gray_img, 0, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_TRIANGLE)
masked = cv2.bitwise_and(image, image, mask=threshInv)
plt.imshow(masked)
print("Threshold: ", T)
```
