### Import Libraries

```python
import cv2
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure, title
import glob
import numpy as np
```

### Define Thresholding Operations

```python
def manual_thresholding(image, threshold):
    _, thresh_msk = cv2.threshold(gray_img, threshold, 255, cv2.THRESH_BINARY)
    masked = cv2.bitwise_and(image, image, mask=thresh_msk)
    return masked
    
def triangle_thresholding(image):
    _, thresh_msk = cv2.threshold(gray_img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_TRIANGLE)
    masked = cv2.bitwise_and(image, image, mask=thresh_msk)
    return masked

def mean_adaptive_thresholding(image):
    thresh_msk = cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 11, 8)
    masked = cv2.bitwise_and(image, image, mask=thresh_msk)
    return masked

def gaussian_adaptive_thresholding(image):
    thresh_msk = cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 19, 6)
    masked = cv2.bitwise_and(image, image, mask=thresh_msk)
    return masked
```

### Apply thresholds to Images

```python
cnt = 1
m_thresh = 100
x_dim = 5
y_dim = 1
img = "../input/doc.jpeg"
figure(figsize=[8,30])
plt.subplot(x_dim,y_dim,cnt)
image = plt.imread(img)
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
title('Original Image')
plt.imshow(image, aspect='auto')
cnt+=1
plt.subplot(x_dim,y_dim,cnt)
title('Manual Thresholding')
plt.imshow(manual_thresholding(image, m_thresh), aspect='auto')
cnt+=1
plt.subplot(x_dim,y_dim,cnt)
title('Triangle Thresholding')
plt.imshow(triangle_thresholding(image), aspect='auto')
cnt+=1
plt.subplot(x_dim,y_dim,cnt)
title('Mean Adaptive Thresholding')
plt.imshow(mean_adaptive_thresholding(image), aspect='auto')
cnt+=1
plt.subplot(x_dim,y_dim,cnt)
title('Gaussian Adaptive Thresholding')
plt.imshow(mean_adaptive_thresholding(image), aspect='auto')
cnt+=1
```
